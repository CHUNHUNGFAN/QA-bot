```
#Midterm Version 3
#Author: CHUN HUNG FAN
#Date:2019.01.14
#主程式

import json
import jieba
import jieba.posseg as pseg
import time

tStart = time.time()
answerlist = []
answerlistno = []
jieba.set_dictionary("dict.txt.big")
jieba.load_userdict("/content/drive/My Drive/midterm/dict_big.txt")


with open('/content/drive/My Drive/midterm/10_Question.json', 'r') as f:
  JsonList = json.load(f)

q_index_start = 0
q_index_end = 200

ite = q_index_start
for idx1 in range(q_index_start,q_index_end):
  tmpResult = []
  tmpResult2 = []
  questionindex = set()
  question = JsonList[idx1]['Question']
  a_A = JsonList[idx1]['A']
  l_A = len(a_A)
  a_B = JsonList[idx1]['B']
  l_B = len(a_B)
  a_C = JsonList[idx1]['C']
  l_C = len(a_C)
  
  if a_A in question:
    answerlist.append("A")
    answerlistno.append("A")
  elif a_B in question:
    answerlist.append("B")
    answerlistno.append("B")
  elif a_C in question:
    answerlist.append("C")
    answerlistno.append("C")
  else:
    w_n = "wc" + str(l_A)
    w_n = globals()[w_n]
    if a_A in w_n:
      tmpResult.append("O")
    else:
      tmpResult.append("X")

    w_n = "wc" + str(l_B)
    w_n = globals()[w_n]
    if a_B in w_n:
      tmpResult.append("O")
    else:
      tmpResult.append("X")

    w_n = "wc" + str(l_C)
    w_n = globals()[w_n]
    if a_C in w_n:
      tmpResult.append("O")
    else:
      tmpResult.append("X")

    if tmpResult.count("O") == 1:
      idx = tmpResult.index("O")
      if idx == 0:
        answerlist.append(str(idx1)+"A")
        answerlistno.append("A")
      elif idx == 1:
        answerlist.append(str(idx1)+"B")
        answerlistno.append("B")
      else:
        answerlist.append(str(idx1)+"C")
        answerlistno.append("C")
    else:
      answerindex_A = set()
      answerindex_B = set()
      answerindex_C = set()

      order = [i for i, v in enumerate(tmpResult) if v == "O"]

      for idx3 in range(len(order)):
        if order[idx3] == 0:
          answer_word =  a_A
          a_i = "wi" + str(l_A)
          a_i = globals()[a_i]
          if answer_word in a_i:
            a_index = a_i.index(answer_word)
            a_index += 1
            while a_i[a_index] != u'':
              answerindex_A.add(a_i[a_index])
              a_index += 1
        elif order[idx3] == 1:
          answer_word =  a_B
          b_i = "wi" + str(l_B)
          b_i = globals()[b_i]
          if answer_word in b_i:
            b_index = b_i.index(answer_word)
            b_index += 1
            while b_i[b_index] != u'':
              answerindex_B.add(b_i[b_index])
              b_index += 1
        elif order[idx3] == 2:
          answer_word =  a_C
          c_i = "wi" + str(l_C)
          c_i = globals()[c_i]
          if answer_word in c_i:
            c_index = c_i.index(answer_word)
            c_index += 1
            while c_i[c_index] != u'':
              answerindex_C.add(c_i[c_index])
              c_index += 1

      questionNew =""
      questionIndexList = []


      words = pseg.cut(question)
      for word in words:
        questionNew += word.word + " " + word.flag +"/"
      spli = questionNew.split('/')
      for i in range(len(spli)):
        wordsplit = spli[i].split(" ")
        l = len(wordsplit)
        if 'n' in wordsplit[l-1]:
          questionIndexList.append(wordsplit[0]) 

      for idx2 in range(len(questionIndexList)):
        l_q_t = len(questionIndexList[idx2])
        q_i = "wi" + str(l_q_t)
        q_i = globals()[q_i]
        if questionIndexList[idx2] in q_i:    
          n_q_i = q_i.index(questionIndexList[idx2])
          n_q_i += 1
          for idx4 in range(200):
            if q_i[n_q_i] != u'':
              questionindex.add(q_i[n_q_i])
              n_q_i += 1
            else:
              break

      a_l_A = len(answerindex_A & questionindex)
      a_l_B = len(answerindex_B & questionindex)
      a_l_C = len(answerindex_C & questionindex)
      tmpResult2.append(a_l_A)
      tmpResult2.append(a_l_B)
      tmpResult2.append(a_l_C)

      dic = {str(a_l_A):'A',str(a_l_B):'B',str(a_l_C):'C'}
      maxi = max(a_l_A,a_l_B,a_l_C)
      midi = min(max(a_l_A,a_l_B),max(a_l_B,a_l_C),max(a_l_A,a_l_C))

      if a_l_A == a_l_B == a_l_C == 0:
        answerlist.append(str(ite)+dic[str(maxi)])
        answerlistno.append(dic[str(maxi)])
        print "請查看:第"+str(ite)+"題"
        print question
        print "A:",a_A
        print "B:",a_B
        print "C:",a_C
        print("AI's Answer : " + dic[str(maxi)])
        print('\n')

      elif tmpResult2.count(0) == 2:
        d = max(a_l_A,a_l_B,a_l_C)
        answerlist.append(dic[str(maxi)])
        answerlistno.append(dic[str(maxi)])

      elif abs(maxi- midi < 12):
        answerlist.append(str(ite)+dic[str(maxi)])
        answerlistno.append(dic[str(maxi)])
        print "請查看:第"+str(ite)+"題"
        print question
        print "A:",a_A
        print "B:",a_B
        print "C:",a_C
        print("AI's Answer : " + dic[str(maxi)])
        print('\n')

      else:
        d = max(a_l_A,a_l_B,a_l_C)
        answerlist.append(dic[str(maxi)])
        answerlistno.append(dic[str(maxi)])
  ite +=1
    
print json.dumps(answerlist)
print json.dumps(answerlistno)
    
tEnd = time.time()#計時結束
#列印結果
print "It cost %f sec" % (tEnd - tStart)#會自動做近位
      
```
